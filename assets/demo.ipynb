# Demo


`nb2bear` is a package for adding Jupyter notebooks to Bear. It is built using the [nbdev](https://nbdev.fast.ai) package and written using principles learnt by the author in the Udemy course [Clean Code](https://www.udemy.com/course/writing-clean-code/).

Before adding the notebook, `nb2bear` formats the document which includes:
* Remove additional spaces generated by `nbconvert`'s exporter.
* Convert code cell output to quoted text.
* Convert indentations from whitespace to tab.
* Add a blank line before a heading.

Here's a quick example to show the package's output of a code block. Note that due to Bear's API limitations and the sandboxed nature of Mac apps, the package cannot import images.
